```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

load data

```python
df = pd.read_csv("cust_data.csv")
df
```

```python
tenure_churn_no = df[df.Churn==0].tenure
tenure_churn_yes = df[df.Churn==1].tenure

plt.xlabel("tenure")
plt.ylabel("Number Of Customers")
plt.title("loyal customers")


plt.hist([tenure_churn_yes, tenure_churn_no], rwidth=0.95, color=['red','green'],label=['leaving','not leaving'])
plt.legend()
```

```python
mc_churn_no = df[df.Churn==0].MonthlyCharges      
mc_churn_yes = df[df.Churn==1].MonthlyCharges      

plt.xlabel("Monthly Charges")
plt.ylabel("Number Of Customers")
plt.title("Left Because of monthly charges")


plt.hist([mc_churn_yes, mc_churn_no], rwidth=0.95, color=['red','green'],label=['leaving',' not leaving'])
plt.legend()
```
